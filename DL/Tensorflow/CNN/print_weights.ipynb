```python
import tensorflow as tf
#tf.random_normal,给出均值为mean，标准差为stdev的高斯随机数（场）
a = tf.Variable(tf.random_normal([3, 3, 1, 64], stddev=0.1))
print (a)
a_print = tf.Print(a, [a], "a: ")
#Variable的初始化
init = tf.global_variables_initializer()
#建立会话
sess = tf.Session()
#执行初始化
sess.run(init)
sess.run(a_print)
```
