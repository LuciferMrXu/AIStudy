## 推荐算法类别总结

1. 协同过滤推荐: 常见的技术基于用户的协同过滤、基于物品的协同过滤、基于模型的协同过滤；(必须掌握)
2. 基于内容的推荐：通过自然语言等相关的NLP知识(Word2Vec)，进行数据挖掘，找出相似物品，从而实现推荐。代表算法：基于序列的推荐算法(加分项)
3. 基于知识的推荐系统：基于开发人员定义的用户需求和物品特征之间的映射关系来做推荐，将和用户需要比较相似的物品推荐给用户，这个过程比较类似搜索系统(可选了解)
4. 基于人口信息统计的推荐：类似UserCF，但是区别是计算用户相似度的时候使用的是用户本身的特征属性来计算的(可选了解)
5. 混合推荐：将多个推荐算法进行融合，原理和机器学习中的集成算法一样；(必须掌握)

## 混合推荐

将多种推荐算法模型产生的推荐列表进行融合的一种方式，或者说将多种推荐算法融合的一种方式；常用混合推荐是并行式混合推荐；

假设现在有三个推荐模型，对于同一个用户u，现成产生的推荐列表分别是：<br/>
model1: item1:4.5, item2:4.3, item4:4, item6:3.8, item8:3.8<br/>
model2: item2:4.8, item6:4.1, item3:4, item1:3.8, item9:3.6<br/>
model3: item1:4.9, item4:4.3, item8:3.6, item2:3.2, item10:3.1<br/>

对于model1、model2、model3的结果，可以认为都是在实际评分值附近的，那么我们可以以model1、model2、model3的预测评分作为特征矩阵X,以实际的评分作为Y，训练出一个X->Y的映射函数, 从而可以得到一个最终的推荐模型，可以认为最终的这个融合的模型其实就是普通的回归算法，一般为线性回归

也就是在这里需要将model1、model2、model3的结果作为X，实际值作为Y训练一个最终的融合模型model4，model4输出的是对物品的评分，而这个评分从大到小进行排序后，作为最终的推荐结果

# 序列建模推荐

NLP重点是将文本转换为词向量/特征向量；前提的技术：分词(jieba)；转换的技术：哑编码、词袋法、TF-IDF、Word2Vec；Word2Vec简单来讲：将词转换为词向量，并且在转换之后保留单词之间的关联性，同样意义的单词转换得到的词向量比较接近，并且会将同时出现的单词的词向量也是比较接近

我是小明 => 我 是 小明<br/>
我 => [1, 0, 0]<br>
是 => [0, 1, 0]<br/>
小明 => [0, 0, 1]<br/>
我是小明 => [1, 0, 0, 0, 1, 0, 0 ,0, 1]<br/>
小明是我 => [0, 0, 1, 0, 1, 0, 1, 0, 0]<br/>

如果我们发现，用户对于物品的访问具有一定的访问路线，那么路线中的各个物品体现的是具有一定的相似度/相关性的；所以我们可以一个歌单中的歌曲其实就是一个访问路径，也就是歌曲歌曲之间存在着某种相似度；如果某两首歌经常性的一起在同一个歌单中，那么表示这两首歌比较相似，从而我们可以将其中一首认为是另外一首的相似歌曲。

_**
Word2Vec本质是将词映射成为一定维度的稠密向量，同时保持词与词之间的关联性<br/><br/>
因为“物以类聚，人以群分” ---> 出现在一起的物/人存在一定的关联性 --> 而Word2Vec就是一种把这种关联性提交取来的方式 ==> 可以考虑使用gensim这个python框架库；安装方式：pip install gensim
**_

```python
import multiprocessing
import random
import gensim
import pickle
```

```python
def parse_playlist_2_sequence(playlist_info, playlist_sequence):
    """
    将输入的歌单信息转换为序列
    :param playlist_info:
    :param playlist_sequence:
    :return:
    """
    # 把歌单id认为就是NLP中的分词后的词
    song_sequence = []
    contents = playlist_info.strip().split("\t")
    for song in contents[1:]:
        try:
            # 获取歌曲数据
            song_id, _, _ = song.split('::::')
            # 添加序列
            song_sequence.append(song_id)
        except Exception as e:
            print(e)
            pass

    # 将歌曲序列添加到歌单序列中，认为歌单中的歌曲打乱顺序也是可以的
    for i in range(len(song_sequence)):
        # 随机打乱顺序
        random.shuffle(song_sequence)
        # 将歌曲序列，加入歌单序列列表中
        playlist_sequence.append(song_sequence)
```

```python
def train_song2vec(in_file, out_file):
    """
    进行模型训练
    :param in_file:
    :param out_file:
    :return:
    """
    # 保存所有歌单序列
    playlist_sequence = []

    # 遍历文件，获取所有歌单序列
    with open(in_file, 'r', encoding='UTF-8') as reader:
        for line in reader:
            parse_playlist_2_sequence(line, playlist_sequence)

    # 使用Word2Vec训练
    cores = multiprocessing.cpu_count()
    print("使用{%d}个CPU进行模型训练!!!" % cores)
    print("开始训练Word2Vec模型....")
    model = gensim.models.Word2Vec(sentences=playlist_sequence, size=100, min_count=3, window=9, workers=cores)
    print("训练Word2Vec模型完成!!!")
    print("开始模型保存...")
    model.save(out_file)
```

```python
def train_predict(model_file_path, song_id2name_file_path):
    """
    进行序列模型测试
    @:param model_file_path
    :return:
    """
    # 1. 加载模型
    model = gensim.models.Word2Vec.load(model_file_path)

    # 2. 获取相似歌曲推荐
    # 2.1 加载歌曲id -> 歌单名称的映射
    print("开始加载歌曲id->歌曲名称的映射....")
    song_id_2_name = pickle.load(open(song_id2name_file_path, 'rb'))
    print("加载歌曲id->歌曲名称的映射完成!!!")
    # 2.2 获取一部分歌曲id
    song_id_list = list(song_id_2_name.keys())[0:200:10]
    # 2.3 对于这部分歌曲获取相似的歌曲
    for song_id in song_id_list:
        # 获取最相似歌曲的10首歌的id
        result_song_list = model.most_similar(song_id, topn=10)

        # 输出信息
        print("\n%s -- %s" % (song_id, song_id_2_name[song_id]))
        print("相似歌曲及相似度分别为:")
        for sim_song_id in result_song_list:
            print("\t%s -- %s  %.5f" % (sim_song_id[0], song_id_2_name[sim_song_id[0]], sim_song_id[1]))
```

```python
# 定义相关变量
music_playlist_song_file_path = "./datas/163_music_playlist.txt"
song_pkl_file_path = "./datas/song.pkl"
sequence_model_file_path = "./datas/sequence.model"
```

```python
# 模型训练
train_song2vec(music_playlist_song_file_path, sequence_model_file_path)
```

```python
# 模型使用/模型预测
train_predict(sequence_model_file_path, song_pkl_file_path)
```
