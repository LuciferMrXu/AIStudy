## 协同过滤简单总结(CF)

是推荐系统中比较经典，而且使用比较广泛的一种算法；包含两个部分：线上的实时协同和线下离线的过滤；协同指的就是实时的找到用户可能喜欢的商品作为推荐列表，过滤一些不可能推荐的商品数据，比如过滤评分比较低的数据，或者过滤虽然评分高但是已经购买过的商品。

主要分为三种:
1. 基于用户的协同过滤 -> UserCF
2. 基于物品的协同过滤 -> ItemCF
3. 基于模型的协同过滤 -> ModelCF

## 基于模型(Model-Based)的协同过滤
主要思想：利用机器学习的思路来建立推荐系统的模型

### 1. 利用关联规则做协同过滤
找出N项的频繁项集，然后就可以认为当用户购买这个频繁项集中的某一个商品u后，同时购买其它商品的概率是最大的，那么当用户在浏览某个商品u的时候，我们就可以将这个商品的对应的关联的其它商品(N项频繁项集中的其它商品)推荐给当前用户，这个就是类似亚马逊或者当当上的"一起购买"那个地方的展示信息。
### 2. 基于隐语义模型/基于矩阵分解做协同过滤
通过将原始评分矩阵进行分解，从而得到得到最终的预测模型
### 3. 使用聚类算法做协同过滤
思想和UserCF/ItemCF类似，我们可以按照用户/物品的特性基于一定的距离公式进行聚类。如果是基于用户聚类，可以将用户按照一定的距离度量公式分成不同的人群，将相同群体中评分比较高的物品推荐给当前群体的其它用户。
### 4. 使用分类算法做协同过滤
以用户id和物品id作为特征属性矩阵(一般会加入物品的相同特性的特征属性)，然后将用户对于物品的评分进行区间化(将用户评分按照高低，分成几个区间)作为y值, 转换为分类算法；一般情况下，我们使用二分类比较多，一般情况下使用logistic回归比较多。
### 5. 使用回归算法做协同过滤
直接构建一个回归算法模型，来训练出一个能够得到目标用户对于某个物品的评分的这样一个模型。
### 6. 使用深度学习的思想做协同过滤
核心：将原来一个隐藏模型的这种形式转换为多隐藏的模型来做推荐。
