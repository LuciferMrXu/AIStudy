```python
import pandas as pd
import matplotlib.pyplot as plt
admissions = pd.read_csv("admissions.csv")
print(admissions.head())
plt.figure()
plt.scatter(admissions['gpa'], admissions['admit'])
plt.show()
```

```python
import numpy as np

# 自己写Logit函数
def logit(x):
    # np.exp(x)将x提高到指数幂，即e^x.e~=2.71828。
    return np.exp(x)  / (1 + np.exp(x)) 
    
# 在-6和6之间生成50个均匀间隔的实数值（等差数列）。
x = np.linspace(-6,6,50, dtype=float)

#使用logit函数转换t中的每个数字（权重参数）。
y = logit(x)

# 画图展示数据
plt.plot(x, y)
plt.ylabel("Probability")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(admissions[["gpa"]], admissions["admit"])
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()
logistic_model.fit(admissions[["gpa"]], admissions["admit"])
```

```python
logistic_model = LogisticRegression()
logistic_model.fit(admissions[["gpa"]], admissions["admit"])
pred_probs = logistic_model.predict_proba(admissions[["gpa"]])  # 预测概率值
plt.scatter(admissions["gpa"], pred_probs[:,1])  # 填1纵轴为被接受的概率，填0纵轴为被拒绝的概率
plt.show()
```

```python
logistic_model = LogisticRegression()
logistic_model.fit(admissions[["gpa"]], admissions["admit"])
fitted_labels = logistic_model.predict(admissions[["gpa"]])   # 准确类别预测
plt.scatter(admissions["gpa"], fitted_labels)
plt.show()
```
