# matplotlib：刻度、标题、标签和图例

```python
'''
legend ()：生成默认图例, matplotlib 中的 legend 图例就是为了帮我们展示出每个数据对应的图像名称. 更好的让读者认识到你的数据结构.
xlabel、ylabel:设置X轴Y轴标签 
title：设置标题
xlim、ylim：控制图标的范围
xticks、yticks: 控制图标的刻度
gca获取当前坐标轴信息。使用spines设置边框,使用set_color设置边框颜色：默认白色
'''
x1=[1,2,3]
y1=[5,7,4]
x2=[1,2,3]
y2=[10,14,12]
# 设置线条标签
plt.plot(x1,y1,'ro--',label='First Line')
plt.plot(x2,y2,'b-',label='Second Line')
# 设置标题、标签
plt.xlabel('月份')    # 设置x轴标签
plt.ylabel('年份')  # 设置y轴标签
plt.title('进出口数据')   # 标题
# 设置x轴范围
plt.xlim(1,3)
# 设置y轴范围
plt.ylim(0,15)
# 设置x轴刻度
plt.xticks(np.linspace(0,6,5))
# 设置y轴刻度
plt.yticks(np.arange(1,15,3),['2011年','2012年','2013年','2014年','2015年'])
# 获取坐标轴信息
ax=plt.gca()
# 设置边框
ax.spines['right'].set_color(None)
ax.spines['top'].set_color(None)
# lengend 生成默认图例
plt.legend()
plt.show()
```

# Matplotlib：Subplot 子图

```python
x1=[1,2,3]
y1=[5,7,4]
x2=[1,2,3]
y2=[10,14,12]
plt.subplot(221)
plt.plot(x1,y1,'r-')
plt.subplot(223)
plt.plot(x2,y2,'b--')
plt.show()
```

```python
x=np.arange(-3,3,0.1)
y1=np.sin(x)
y2=np.cos(x)

plt.figure()
plt.subplot(221)
plt.plot(x,y2)
plt.plot(x,y1)

plt.subplot(222)
plt.plot(x,y2)
plt.plot(x,y1)

plt.subplot(223)
plt.plot(x,y2)
plt.plot(x,y1)

plt.subplot(224)
plt.plot(x,y2)
plt.plot(x,y1)

plt.show
```

```python
# 获取figure对象
fig=plt.figure(figsize=(8,6))
# 在figure上创建axes对象
ax1=fig.add_subplot(221)
ax2=fig.add_subplot(222)
ax3=fig.add_subplot(223)
# 在ax1上绘图
ax1.plot(np.random.randn(50).cumsum(),'g-')
# 在ax2上绘图
ax2.plot(np.random.randn(50).cumsum(),'b--')
# 在ax3上绘图
ax3.plot(np.random.randn(50).cumsum(),'k--')

plt.show()
```

```python
# 面向对象的写法
fig=plt.figure()
# 在figure上创建对象
ax1=fig.add_subplot(321)
ax2=fig.add_subplot(322)
ax3=fig.add_subplot(312)
ax4=fig.add_subplot(325)
ax4=fig.add_subplot(326)
# 绘图
ax1.plot(np.random.randn(50).cumsum(),'g--')
ax2.plot(np.random.randn(50).cumsum(),'b--')
ax3.plot(np.random.randn(50).cumsum(),'m--')
ax4.plot(np.random.randn(50).cumsum(),'g--')
ax4.plot(np.random.randn(50).cumsum(),'b--')

plt.suptitle('subplot')
plt.show()
```

# Subplots

```python
'''
Sublots():返回一个图像和多个子图
参数：  nrows=x, ncols=x, sharex=True, sharey=False, gridspec_kw={'height_ratios':[2,2,1,1]}
例：
fig, ax = plt.subplots(2,2),其中参数分别代表子图的行数和列数，一共有 2x2 个图像。函数返回一个figure图像和一个子图ax的array列表。
'''
Plot,axes=plt.subplots(nrows=4, ncols=1, sharex=True, sharey=False)
# 设置fig标题
Plot.suptitle('test',fontsize=20)

axes[0].plot(range(10),'ro-')
axes[1].plot(range(10),'bo-')
axes[2].plot(range(10),'go-')
axes[3].plot(range(10),'mo-')

plt.show()
```

```python
fig,ax=plt.subplots(2,2)
for i in range(2):
    for j in range(2):
        ax[i][j].plot(np.random.randn(10).cumsum())
# 设置子图之间的距离
fig.subplots_adjust(wspace=0,hspace=0)
# 图像保存文件
plt.savefig('a.jpg')
plt.show()
```

# matplotlib曲线图

```python
x=np.linspace(0,10,1000)
y=np.sin(x)
z=np.cos(x**2)
# 创建一个绘图对象，指定宽8英寸，高4英寸
plt.figure(figsize=(8,4))
# 只要在字符串前后添加“$”符号，里面的字符串会自动绘制成数学公式
plt.plot(x,y,label='$sin(x)$',color='red',linewidth=2)
plt.plot(x,z,'b--',label='$cos(x^2)$')
plt.xlabel('Time(s)')
plt.ylabel('Volt')
plt.title('PyPlot First Example')
# 设置y轴范围
plt.ylim(-1.2,1.2)
# 显示图示说明
plt.legend()
# 显示网格
plt.grid(True)
plt.show()
```

# Matplotlib柱状图

```python
'''
matplotlib.pyplot. bar (*args, **kwargs)
bar(left, height, width, bottom, * args, align='center', **kwargs)
参数： 
left:数据标量 
height：高 width:款 
bottom：底端对应Y轴
align:对齐如果为 "居中", 则将x参数解释为条形中心的坐标。如果 "边缘", 将条形按其左边缘对齐要对齐右边缘的条形图, 可传递负的宽度和对align='edge'
'''
plt.figure()
plt.bar([1,3,5,7,9,11],
        [5,2,7,8,2,6],
        label='Example one',color='y'
)
plt.bar([2,4,6,8,10,12],
        [8,6,2,5,6,3],
        label='Example two',color='g'
)
plt.legend()
plt.xlabel('bar number')
plt.ylabel('bar height')

plt.title('举个栗子')
plt.show()
```

```python
# 设置柱子数量
n=3
# 设置宽度
width=0.5/n
# 柱子组数
left=np.arange(1,5)
height=np.array([200,300,400,500])

plt.figure()
plt.bar(left-width,height,width=width,color='g',align='center',label='green')    
plt.bar(left,height*3,width=width,color='y',align='center',label='yellow')
plt.bar(left+width,height*2,width=width,color='b',align='center',label='blue')

plt.xticks(np.arange(1,5),[str(i)+'月' for i in range(1,6)]) 

plt.xlabel('月份')
plt.ylabel('美元')

plt.legend()
plt.show()
```

```python
plt.figure()
n=3
width=0.6/n
x=np.arange(1,5)
y=np.random.randint(100,500,4)

plt.bar(x-0.2,height=y,width=width,color='blue',align='center',label='aa')    
plt.bar(x,height=y+50,width=width,color='red',align='center',label='bb')
plt.bar(x+0.2,height=y+100,width=width,color='green',align='center',label='cc')

plt.title('柱状图')
plt.xticks(np.arange(1,5),[str(i)+'月' for i in range(1,5)]) 
plt.yticks(np.arange(100,800,100),[str(i)+'元' for i in range(100,800,100)]) 

plt.xlabel('日期')
plt.ylabel('金额')

ax=plt.gca()
ax.spines['right'].set_color(None)
ax.spines['top'].set_color(None)

plt.legend(loc='upper left')
plt.show()
```

# Matplotlib直方图

```python
'''
matplotlib.pyplot.hist(
	x,bins=10,range=None,normed=False,weights=None,cumulative=False,
	histtype='bar', align='mid',orientation=u'vertical', log=False,color=None, 	label=None,stacked=False,hold=None,**kwargs) 
– x: 一个列表或者多个列表（表示不同数据集，长度可以不一致）
– range: 元组
– weights: x里每个元素对bin高度的贡献（默认为1）
– histtype: ['bar' | 'barstacked' | 'step' | 'stepfilled'] 
– align: ['left' | 'mid' | 'right'] 
– orientation: ['horizontal' | 'vertical'] 
– color: 一种颜色或者颜色列表（针对不同数据集）
'''
# 设置数据
x=np.random.randn(10000).cumsum()
# x=np.random.randn(10000)*10*15
# 绘图
plt.figure()
plt.hist(x,1000,color='red',alpha=0.7)
# 添加网格
plt.grid(True)
plt.show()
```

```python
# 展现出10000个随机数，满足平均值是100，标准差是15，呈现一个正态分布曲线
mu,sigma=100,15
x=mu+np.random.randn(10000)*sigma

plt.hist(x,100,density=1,facecolor='g',alpha=0.75)
plt.title('直方图')
plt.text(60,0.025,'$\mu=100,\sigma=15$')
plt.axis([40,160,0,0.03])   # 设置横纵坐标范围
plt.grid(True)
plt.show()
```

```python
fig,axes=plt.subplots(2,2,sharex=True,sharey=True)
# print(axes)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500),bins=10,color='k',alpha=0.5)
plt.subplots_adjust(wspace=0,hspace=0)    # 调整子图之间的距离
plt.show()
```

# matplotlib绘制散点图

```python
'''
matplotlib.pyplot. scatter (x, y, s=none, c=none, marker=none, cmap= None, 
                            norm=none, vmin=none, vmax=none, alpha=none, Linewidths=none,
                            verts=none, edgecolors=none, hold=none, data= None, **kwargs)
参数：
    x,y:相同长度的数组序列
    s  :散点的大小标量或形同数组，可选参数，默认20
    c  :散点的色彩或颜色序列，可选
    maker:标记风格，可选，默认是‘o’
    alpha:散点的透明度
'''
x=[1,2,3,4,5,6,7,8]
y=[5,2,4,2,1,4,5,2]
T=np.random.randn(8)*125

plt.scatter(x,y,label='散点分布',c=T,s=25,marker='o',alpha=0.5)
plt.title('散点图')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
x=np.random.rand(1000)
y=np.random.randn(1000)

plt.figure()
plt.scatter(x,y,c=np.random.rand(1000),s=np.random.random(1000)*100,alpha=0.8)
plt.title('散点图')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.axis([0,5,0,20])
plt.title('My First Chart',fontsize=20,fontname='Times New Roman')
plt.xlabel('Counting',color='gray')
plt.ylabel('Square values',color='gray')
plt.text(1,1.5,'First')
plt.text(2,4.5,'Second')
plt.text(3,9.5,'Third')
plt.text(4,16.5,'Fourth')
plt.text(1,11.5,'$y=x^2$',fontsize=20,bbox={'facecolor':'yellow','alpha':0.2})
plt.grid(True)
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.plot([1,2,3,4],[0.8,3.5,8,15],'g^')
plt.plot([1,2,3,4],[0.5,2.5,5.4,12],'b*')
plt.legend(['First series','Second series','Third series'],loc=2)
plt.savefig('my_chart.png')
plt.show()
```

# matplotlib绘制饼图

```python
'''
matplotlib.pyplot. pie(x, explode=None, labels=None, colors=('b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'), autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=None, radius=None, counterclock=True, wedgeprops=None, textprops=None, center = (0, 0), frame = False ) 
参数:
    x       (每一块)的比例，如果sum(x) > 1会使用sum(x)归一化
    labels  (每一块)饼图外侧显示的说明文字
    explode (每一块)离开中心距离
    startangle  起始绘制角度,默认图是从x轴正方向逆时针画起,如设定=90则从y轴正方向画起
    shadow  是否阴影
    labeldistance label绘制位置,相对于半径的比例, 如<1则绘制在饼图内侧
    autopct 控制饼图内百分比设置,可以使用format字符串或者format function
            '%1.1f'指小数点前后位数(没有用空格补齐)
    pctdistance 类似于labeldistance,指定autopct的位置刻度
    radius  控制饼图半径
'''
labels=['第一部分','第二部分','第三部分']
explode=(0,0,0.2)
colors=['red','g','y']
# 比例
x=[30,70,20]

plt.figure()
plt.pie(x,explode=explode,labels=labels,colors=colors,startangle=90,
        shadow=True,labeldistance=1.1,autopct='%1.1f%%',pctdistance=0.6)
# 设置饼图为圆
plt.axis('equal')
plt.show()
```

```python
# 设置数据
x=np.array([15,20,25,30,10])
# 设置距离
explode=[0,0,0,0.2,0]
labels=['网络购物','旅游交通','日常生活','女朋友','游戏']
colors=['r','b','g','y','m']
plt.figure()
plt.pie(x,explode=explode,labels=labels,colors=colors,startangle=90,
        shadow=True,labeldistance=1.2,autopct='%.2f%%',pctdistance=0.5)
plt.title('明细')
plt.axis('equal')
plt.show()
```

# Pandas之Series 制图

```python
import pandas as pd
# 单线段
s=pd.Series(np.random.randn(10).cumsum(),index=np.arange(0,100,10))
s.plot()
```

```python
ser=pd.Series(np.random.randn(10),index=np.arange(0,100,10))
ser
plt.plot(ser,color='red')
```

```python
# 多线段
y1=np.array([600,500,800,400,300,500])   # 进口
y2=np.array([500,900,800,650,700,600])   # 出口
ser1=pd.Series(y1,index=['1月','2月','3月','4月','5月','6月'])
ser2=pd.Series(y2,index=['1月','2月','3月','4月','5月','6月'])
ser1.plot(style='g-o',label='鼓浪屿')
ser2.plot(label='张家界')
plt.legend() # 生成图例
```

```python
ser1=pd.Series(np.array([100,600,300,400,500]),index=['1月','2月','3月','4月','5月'])
ser2=pd.Series(np.array([200,300,400,600,200]),index=['1月','2月','3月','4月','5月'])
plt.plot(ser1,color='green',label='一班')
plt.plot(ser2,color='blue',label='二班')
plt.legend()
plt.show()
```

# Pandas之DataFrame制图

```python
y=np.array([[600,500,800,400,300,500],[500,900,800,650,700,600]])
# 行索引对应x轴，列索引对应label
df1=pd.DataFrame(y.T,index=['1月','2月','3月','4月','5月','6月'],columns=['鼓浪屿','张家界'])
df1.plot(style=['r-o','b-o'])
```

```python
df1.plot(kind='bar',style=['r-o','b-o'])
plt.legend()
plt.title('旅游人次')
plt.xlabel('月份')
plt.ylabel('人次/百万')
```
