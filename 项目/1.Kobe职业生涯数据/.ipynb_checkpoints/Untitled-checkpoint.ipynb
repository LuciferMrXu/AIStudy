```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 导入数据
filename= "data.csv"
raw = pd.read_csv(filename)
print (raw.shape)
raw.head()
# raw.tail()  # 打印后五行
```

```python
# 5000 个缺失值（后面会提取出来当作验证集）
kobe =  raw[pd.notnull(raw['shot_made_flag'])]  # notnull提取出值不为空的行
print (kobe.shape)
```

```python
#plt.subplot(211) 第一个数字是行，第二个数字是列
alpha = 0.02
plt.figure(figsize=(10,10))

# loc_x and loc_y 横纵坐标抽
plt.subplot(121)    # 一行两列的第一个子图
plt.scatter(kobe.loc_x, kobe.loc_y, color='R', alpha=alpha)  # α参数表示透明度
plt.title('loc_x and loc_y')

# lat and lon 经度和纬度
plt.subplot(122)   # 一行两列的第二个子图
plt.scatter(kobe.lon, kobe.lat, color='B', alpha=alpha)
plt.title('lat and lon')
```

```python
'''
    转换为极坐标系
'''
raw['dist'] = np.sqrt(raw['loc_x']**2 + raw['loc_y']**2)

loc_x_zero = raw['loc_x'] == 0
# print (loc_x_zero)   通过bool型进行过滤
raw['angle'] = np.array([0]*len(raw))     # 先把angle全部赋值为0，再用bool值筛选
raw.loc[:,'angle'][~loc_x_zero] = np.arctan(raw.loc[:,'loc_y'][~loc_x_zero] / raw.loc[:,'loc_x'][~loc_x_zero])
raw.loc[:,'angle'][loc_x_zero] = np.pi / 2 
print(raw['angle'])
```

```python
# 将原数据中的‘还剩多少分钟结束’和‘还剩多少秒结束’两列整合成一个特征
raw['remaining_time'] = raw['minutes_remaining'] * 60 + raw['seconds_remaining']
```

```python
# unique（）打印某一列一共包含多少种属性
print(kobe.action_type.unique())
print(kobe.combined_shot_type.unique())
print(kobe.shot_type.unique())
# value_counts(）打印某一列中每个属性分别出现了多少次
print(kobe.shot_type.value_counts())
```

```python
kobe['season'].unique()
```

```python
# 将赛季中的‘-’去掉，取右边的部分代表某一赛季
raw['season'] = raw['season'].apply(lambda x: int(x.split('-')[1]) )
raw['season'].unique()
```

```python
print(kobe['team_id'].unique())
print(kobe['team_name'].unique())
```

```python
pd.DataFrame({'matchup':kobe.matchup, 'opponent':kobe.opponent})
```

```python
plt.figure(figsize=(5,5))
# 将前面做的极坐标distance和原数据中的‘shot_distance’做相关性预测
plt.scatter(raw.dist, raw.shot_distance, color='blue')
plt.title('dist and shot_distance')
```

```python
# groupby(),以填进去的列名为key，对其他列进行相应的计算
gs = kobe.groupby('shot_zone_area')
print (kobe['shot_zone_area'].value_counts())
print (len(gs))
```

```python
import matplotlib.cm as cm
plt.figure(figsize=(20,10))

def scatter_plot_by_category(feat):
    alpha = 0.1
    gs = kobe.groupby(feat)
    cs = cm.rainbow(np.linspace(0, 1, len(gs)))
    for g, c in zip(gs, cs):  #通过zip函数将出手位置的6个属性和彩虹图的颜色值一一对应
        plt.scatter(g[1].loc_x, g[1].loc_y, color=c, alpha=alpha)

# shot_zone_area
plt.subplot(131)
scatter_plot_by_category('shot_zone_area')
plt.title('shot_zone_area')

# shot_zone_basic
plt.subplot(132)
scatter_plot_by_category('shot_zone_basic')
plt.title('shot_zone_basic')

# shot_zone_range
plt.subplot(133)
scatter_plot_by_category('shot_zone_range')
plt.title('shot_zone_range')
```

```python
# 删除原数据中已经处理过的或被替换的特征
drops = ['shot_id', 'team_id', 'team_name', 'shot_zone_area', 'shot_zone_range', 'shot_zone_basic', \
         'matchup', 'lon', 'lat', 'seconds_remaining', 'minutes_remaining', \
         'shot_distance', 'loc_x', 'loc_y', 'game_event_id', 'game_id', 'game_date']
for drop in drops:
    raw = raw.drop(drop, 1)
```

```python
print (raw['combined_shot_type'].value_counts())
# get_dummie()将6个不同投篮位置的字符型特征映射成数字特征（onehot编码）
pd.get_dummies(raw['combined_shot_type'], prefix='combined_shot_type')[0:5]  # prefix属性指定列名前缀
```

```python
categorical_vars = ['action_type', 'combined_shot_type', 'shot_type', 'opponent', 'period', 'season']
for var in categorical_vars:
    raw = pd.concat([raw, pd.get_dummies(raw[var], prefix=var)], axis=1)  # contact()按列连接
    raw = raw.drop(var, 1)       # 将拼接前的数据删除
```

```python
# 选出预测特征中没有异常样本的数据当作训练集
train_kobe = raw[pd.notnull(raw.loc[:,'shot_made_flag'])] 
train_label = train_kobe['shot_made_flag']
print(train_label)
train_kobe = train_kobe.drop('shot_made_flag', 1)   # 去label
# 选出预测特征中是异常样本的数据当作测试集
test_kobe = raw[pd.isnull(raw.loc[:,'shot_made_flag'])]
test_kobe = test_kobe.drop('shot_made_flag', 1)    # 去label
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import confusion_matrix,log_loss  # 导入混淆矩阵和损失函
import time
import numpy as np
```

```python
# find the best n_estimators for RandomForestClassifier（最佳决策树个数）
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold

print('Finding best n_estimators for RandomForestClassifier...')
min_score = 100000
best_n = 0
scores_n = []
range_n = np.linspace(100,500,10).astype(int)
for n in range_n:
    print("the number of trees : {0}".format(n))
    t1 = time.time()
    
    rfc_score = 0.
    rfc = RandomForestClassifier(n_estimators=n)
    kf=KFold(n_splits=10,shuffle=True)
    for train_k, test_k in kf.split(train_kobe):
        rfc.fit(train_kobe.iloc[train_k], train_label.iloc[train_k])
        #rfc_score += rfc.score(train.iloc[test_k], train_y.iloc[test_k])/10
        pred = rfc.predict(train_kobe.iloc[test_k])
        rfc_score += log_loss(train_label.iloc[test_k], pred) / 10
        #   通过log_loss交叉熵损失函数来衡量模型的效果      
    scores_n.append(rfc_score)
    if rfc_score < min_score:
        min_score = rfc_score
        best_n = n
        
    t2 = time.time()
    print('Done processing {0} trees ({1:.3f}sec)'.format(n, t2-t1))
print('最佳决策树个数为%s，其交叉熵损失函数为%s'%(best_n, min_score))


# find best max_depth for RandomForestClassifier（最佳决策树深度）
print('Finding best max_depth for RandomForestClassifier...')
min_score = 100000
best_m = 0
scores_m = []
range_m = np.linspace(4,20,10).astype(int)
for m in range_m:
    print("the max depth : {0}".format(m))
    t1 = time.time()
    
    rfc_score = 0.
    rfc = RandomForestClassifier(max_depth=m, n_estimators=best_n)
    for train_k, test_k in kf.split(train_kobe):
        rfc.fit(train_kobe.iloc[train_k], train_label.iloc[train_k])
        #rfc_score += rfc.score(train.iloc[test_k], train_y.iloc[test_k])/10
        pred = rfc.predict(train_kobe.iloc[test_k])
        rfc_score += log_loss(train_label.iloc[test_k], pred) / 10
        #   通过log_loss交叉熵损失函数来衡量模型的效果 
    scores_m.append(rfc_score)
    if rfc_score < min_score:
        min_score = rfc_score
        best_m = m
    
    t2 = time.time()
    print('Done processing {0} trees ({1:.3f}sec)'.format(m, t2-t1))
print('最佳决策树深度为%s，其交叉熵损失函数为%s'%(best_m, min_score))
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.plot(range_n, scores_n)
plt.ylabel('score')
plt.xlabel('number of trees')

plt.subplot(122)
plt.plot(range_m, scores_m)
plt.ylabel('score')
plt.xlabel('max depth')
```

```python
# 利用交叉验证得来的参数构建最优的随机森林模型
model = RandomForestClassifier(n_estimators=best_n, max_depth=best_m)
model.fit(train_kobe, train_label)
train_predict = model.predict(train_kobe)
```

```python
from sklearn.metrics import classification_report
print("各个特征的重要性权重系数（值越大，对应的特征属性就越重要）：{}".format(model.feature_importances_))
print("训练集上的效果(分类评估报告):\n{}".format(classification_report(train_label,train_predict)))
```

```python
test_predict = model.predict(test_kobe)
print('验证集预测结果：',test_predict)
print("返回的预测概率值:\n{}".format(model.predict_proba(test_kobe)))
```
