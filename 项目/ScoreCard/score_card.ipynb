# 变量筛选


```python
import numpy as np
def func(df):
    if df['margin_bad'] == 0:
        df['woe']= -5
    elif df['margin_good'] == 0:
        df['woe']= 5
    else:
        df['woe']= np.log(df['margin_bad']/df['margin_good'])
    return df

# 计算IV值
def cal_IV(df,feature,target,mode=2):
    lst = []
    cols = ['var','value','all','bad']
    # 对bin_feature字段中的所有取值进行iv值计算，再求和
    for val in df[feature].unique():
        # print(val)
        # 统计feature，feature_value,该value导致target=1的个数
        val_num = df[df[feature]==val][feature].count()
        val_bad_num = df[(df[feature]==val) & (df[target]==1)][feature].count()
        lst.append([feature,val,val_num,val_bad_num])
    # print(lst)
    data = pd.DataFrame(lst,columns=cols)
    data['good'] = data['all'] - data['bad']
    data['margin_bad'] = data['bad']/data['bad'].sum()
    data['margin_good'] = data['good']/data['good'].sum()
    # 如果出现inf，可能是分子或分母为0
    # 方法一，woe分子分母加上一个小值（人为引入了噪音误差）
    if mode == 1:
        data['woe'] = np.log((data['margin_bad']+0.01)/(data['margin_good']+0.01))
    # 方法二，直接给分子分母为0的情况进行修正（推荐）
    elif mode ==2:
        data = data.apply(func,axis=1)
    data['iv'] = data['woe']*(data['margin_bad'] - data['margin_good'])
    return data
        
data = cal_IV(df_train,'bin_age','SeriousDlqin2yrs')
data
```

```python
# 统计分箱字段,进行变量筛选
bin_cols = [x for x in df_train.columns.values if x.startswith('bin_')]
bin_cols
```

```python
feature_cols = []
df_woe = pd.DataFrame(columns=['var','value','all','bad','good','margin_bad','margin_good','woe','iv'])
for col in bin_cols:
    data = cal_IV(df_train,col,'SeriousDlqin2yrs',mode=2)
    if data['iv'].sum() > 0.1:
        print(col,data['iv'].sum())
        df_woe = df_woe.append(data,ignore_index=True)
        feature_cols.append(col)
```

```python
df_woe = df_woe[['var','value','woe']]
df_woe.columns=['feature','bin','woe']
df_woe
```

# 评分卡模型转换

```python
'''
    计算每个人的score
    score =A - B * (w0 + w1x1 + ... + w10x10)
    A = 650
    B = 72.13
    A是人为设置的
    B是通过score公式计算出来的（假设翻倍分数PDO为50分）
'''
A = 650
B = 72.13
def generate_scorecard(model_coef,binning_df,features,B):
    lst = []
    cols = ['Variable','Binning','Score']
    # 模型系数
    coef = model_coef[0]
#     print(coef)
    for i in range(len(features)):
        f = features[i]
#         print(f)
        # 得到该feature的WOE规则
        df = binning_df[binning_df['feature']==f]
#         print(df)
        for index,row in df.iterrows():
            # 这里对评分卡的打分值取整，方便后面统计 
            lst.append([f,row['bin'],int(round(-coef[i]*row['woe']*B))])
    data = pd.DataFrame(lst,columns=cols)
    return data

# df_bin_to_woe 规则 bin ==> woe值
score_card = generate_scorecard(model.coef_,df_woe,feature_cols,B)
score_card
```

```python
# 对于每个字段按照降序排序，查看规则
sort_score_card = score_card.groupby('Variable').apply(lambda x:x.sort_values('Score',ascending=False))
sort_score_card
```

```python
def map_to_score(df,score_card):
    scored_columns = list(score_card['Variable'].unique())
    score = 0
    # 分别计算每个分箱特征的分数
    for col in scored_columns:
        feature_to_bin = score_card[score_card['Variable']==col]
        feature_value = df[col]
        temp_score = score_card[score_card['Binning']==feature_value]['Score'].iloc[0]
        score = score+temp_score
    return score

def calculate_score_with_card(df,score_card,A):
    # 计算某一个人五个特征的woe分数的加和
    df['score'] = df.apply(map_to_score,args=(score_card,),axis=1)
    # 再加上基准分A
    df['score'] = df['score']+A
    df['score'] = df['score'].astype(int)
    return df
```
