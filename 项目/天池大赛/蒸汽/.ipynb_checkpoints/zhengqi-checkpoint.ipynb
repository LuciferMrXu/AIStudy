```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import accuracy_score,mean_squared_error#计算roc和auc
from sklearn.tree import DecisionTreeRegressor
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("./datas/zhengqi_train.txt",sep="\t")
x = df.drop(['target'],1)
y = df['target']
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8)
df.head()
```

```python
#构建adaboost模型
tree = DecisionTreeRegressor(max_depth=10)
ada = AdaBoostRegressor(tree,
                         n_estimators=180, learning_rate=1.0)

ada.fit(x_train, y_train)
y_test_pred = ada.predict(x_test)
y_train_pred = ada.predict(x_train)

print("train mse",mean_squared_error(y_train,y_train_pred))
print("test mse", mean_squared_error(y_test,y_test_pred))
```

```python
online_df = pd.read_csv("./datas/zhengqi_test.txt",sep="\t")
online_y = ada.predict(online_df)
#这个写入会导致最后多一行的空行
with open("./datas/online_y.txt",'w') as f:
    for line in online_y:
        print(line)
        f.write(str(line)+"\n")
```

```python
'''
    比赛小技巧，利用线上测试集去制造我们的线下测试集以提高预测准确率
    （通过一轮预测，使两个测试集的数据差不多）
'''
outline=pd.read_csv("./zhengqi_train.txt",sep="\t")
online=pd.read_csv("./zhengqi_test.txt",sep="\t")
outline.head()
```

```python
# 把outline里的y删掉
outline.drop(['target'],1,inplace=True)
```

```python
# 给outline里的数据赋值为0,online为1
outline['label']=0
online['label']=1
df=pd.concat([outline,online])
print('outline',outline.shape)
print('online',online.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
x=df.drop(['label'],1)
y=df['label']
```

```python
# 做一个分类模型的预测结果
lr=LogisticRegression(solver='liblinear')
lr.fit(x,y)
print(lr.predict(outline.drop(['label'],1)))
# 线下集属于线上集的预测概率
outproba1=lr.predict_proba(outline.drop(['label'],1))[:,1]
print(sum(outproba1>0.01))  # 通过调整概率值筛选数据
```

```python
# 用带有predict_proba接口的模型先做第一轮预测，找出与线上测试集数据长得最相似的线下数据
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x,y)
print(knn.predict(outline.drop(['label'],1)))
# 线下集属于线上集的预测概率
outproba2=knn.predict_proba(outline.drop(['label'],1))[:,1]
print(sum(outproba2>0.001))  # 通过调整概率值筛选数据
```

```python
# 取出最像线上测试集的outline数据作为测试集
print(outline[outproba1>0.01].shape)
outline=pd.read_csv("./zhengqi_train.txt",sep="\t")
new_test=outline[outproba1>0.01]
new_train=outline[outproba1<=0.01]
```

```python
x_train=new_train.drop(['target'],1)
y_train=new_train['target']
x_test=new_test.drop(['target'],1)
y_test=new_test['target']
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbdt=GradientBoostingRegressor(learning_rate=0.05,
                               n_estimators=300,
                              subsample=0.9,     # subsample下采样参数，对传入参数进行二次采样
                               max_depth=3
                              )
gbdt.fit(x_train,y_train)
y_test_pred=gbdt.predict(x_test)
y_train_pred=gbdt.predict(x_train)

print('train_mse:',mean_squared_error(y_train,y_train_pred))
print('test_mse:',mean_squared_error(y_test,y_test_pred))
```
